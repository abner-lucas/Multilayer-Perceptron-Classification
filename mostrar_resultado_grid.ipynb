```python
import pandas as pd

datasets = ['AND','OR','XOR','Tabela_Lógica']
name_dt = datasets[3]

df = pd.read_csv('grid_hould_out_'+name_dt+'_3C.csv')
df = df.drop(columns=['Neurônios_saída', 'Erro_max', 'Acertos', 'Erros'])
#df.columns = [f' __________' for col in df.columns]
df.index = [f'' for col in df.index]
df['EMQ'] = df['EMQ'].map(lambda x: f'{x:.9f}')
display(df)
```
